````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
````

````python
def formato_grafica(titulo, ejex, ejey, leyenda=False, xlim=[None, None], ylim=[None, None]):
    plt.rcParams['axes.axisbelow'] = True

    plt.title(titulo, fontsize=15)
    plt.ylabel(ejey, fontsize=13)
    plt.xlabel(ejex, fontsize=13)

    plt.tick_params(direction='out', length=5, width=0.75, grid_alpha=0.3)
    plt.xticks(rotation=0)
    plt.minorticks_on()
    plt.ylim(ylim[0], ylim[1])
    plt.xlim(xlim[0], xlim[1])
    plt.grid(True)
    plt.grid(visible=True, which='major', color='grey', linestyle='-')
    plt.grid(visible=True, which='minor', color='lightgrey', linestyle='-', alpha=0.2)
    
    if leyenda == True:
        plt.legend(loc=True)
    
    plt.tight_layout;
````

Considere la siguiente función:

````python
def heat_tranfer(Ts, Tinf, Ti, A, Rtot, h, eps, sigma=5.67e-8):   # SI units
    return (Tinf - Ti)/(A*Rtot) - (h+eps*sigma*(Ts+Tinf)*(Ts**2+Tinf**2))*(Tinf-Ts)
````

![Root Finding Heat Transfer Example.png](<attachment:Root Finding Heat Transfer Example.png>)

Si revisaron los videos lección de esta semana, sabrán que esta función tiene una raíz en el intervalo $[285, 295]$ y que se refiere a la ecuación con la cual podemos determinar la temperatura de una superficie. La función se ve así:

````python
Tlist = np.linspace(283.15, 308.15, num=1000)

plt.plot(Tlist, heat_tranfer(Tlist,283.15,308.15,1.8,0.25,2,.95), label="Heat Transfer function") # SI units
plt.plot(Tlist, Tlist*0,"--", label="y=0")
plt.legend()
````

A diferencia de ejemplos más sencillos que hemos trabajado, esta función ```heat_transfer``` requiere que ingresemos más de un argumento. 

Para graficarla hemos utilizado ```heat_tranfer(Tlist, 283.15, 308.15, 1.8, 0.25, 2, .95)```, donde Tlist es nuestra variable independiente, y los demás argumentos son los parámetros de la función ```Tinf = 283.15, Ti = 308.15, A = 1.8, Rtot = 0.25, h = 2, eps = 0.95```. 

El argumento ```sigma = 5.67e-8``` no es necesario, ya que está predefinido en la función.
